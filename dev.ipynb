```python
import openai
from PIL import Image
import requests
from io import BytesIO
import os
```

```python
openai_api_key = os.environ["OPENAI_API_KEY"]
prompt = "Funny cat in a hurry trying to finish coding a project before the deadline. "
size="256x256"
```

```python
import tempfile

response = openai.Image.create(
    api_key=openai_api_key,
    prompt=prompt,
    size=size
)
with tempfile.NamedTemporaryFile(suffix=".jpg") as f:
    f.write(requests.get(response["data"][0]["url"]).content)
    display(Image.open(f.name))
```

```python
model = "gpt-3.5-turbo"
messages = [
    {"role": "user", "content": "What is the best guitar?"},
]
max_tokens = 512
```

```python
steam = openai.ChatCompletion.create(
        api_key=openai_api_key,
        model=model,
        messages=messages,
        top_p=1,
        n=1,
        max_tokens=max_tokens,
        temperature=1,
        presence_penalty=0,
        frequency_penalty=0,
        logit_bias={},
        stream=True,
)
```
